```python
sdata = {"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000}
obj3 = pd.Series(sdata) # series created from a dictionary, the keys become the index in the resulting Series
obj3
```

```python
obj3.to_dict() # we can convert a Series back to a dictionary using the to_dict() method
```

```python
states = ["California", "Ohio", "Oregon", "Texas"]
obj4 = pd.Series(sdata , index = states)
obj4
```

```python
obj4.isna() # or pd.isna(ibj4) ; the isna() method can be used to find which values are missing in the Series
```

```python
obj3+obj4 # when we perform operations between Series, if any index pairs are not the same, the respective index will be filled with NA in the result
```

```python
obj4.name = "population"
obj4.index.name = "state"
obj4
```
