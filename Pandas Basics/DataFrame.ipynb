```python
import numpy as np 
import pandas as pd 
from pandas import Series, DataFrame
```

## DataFrame

```python
data = {"state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
"year": [2000, 2001, 2002, 2001, 2002, 2003],
"pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data) # we can only use DataFrame instead of pd.DataFrame because we have imported it from pandas
frame
```

```python
frame.head() #selects only the first 5 rows of the data frame
```

```python
frame.tail() #selects only the last 5 rows of the data frame
```

```python
pd.DataFrame( data , columns =[ "year" , "state" , "pop"]) # we can also specify the order of the columns by using the columns parameter. If we do not specify the order, it will be in the order of the keys in the data dictionary. If we specify a column that is not in the data dictionary, it will be filled with NaN values.
```

```python
frame.columns # to get the column names of the data frame
```

```python
frame["year"]
```

```python
frame2 = pd.DataFrame(data, columns=["year", "state", "pop", "debt"]) # if we add a column that is not in the data dictionary, it will be filled with NaN values.
frame2
```

```python
frame2.loc[1]
```

```python
frame2["debt"] = 16.5 # we can also add a column to the data frame by using the column name as the key and the values as the value. If we add a column that is not in the data dictionary, it will be filled with NaN values.
frame2
```

```python
frame2["debt"] = np.arange(6.)
frame2
```

```python
frame2["eastern"] = frame2["state"] == "Ohio"
frame2
```

```python
del frame2["eastern"] # we can also delete a column from the data frame by using the del keyword and the column name as the key. If we delete a column that is not in the data frame, it will raise a KeyError.
frame2
```

```python
populations = {"Ohio": {2000: 1.5, 2001: 1.7, 2002: 3.6},
    "Nevada": {2001: 2.4, 2002: 2.9}}
frame3 = pd.DataFrame(populations)
frame3
```

```python
frame3.T # to transpose the data frame
```

```python
frame3 
```

```python
frame3.index.name = "year"
frame3.columns.name = "state"
frame3
```

```python
frame3.to_numpy() # to convert the data frame to a numpy array. The index and column labels are not included in the output.
```

### Some Index methods and properties

| Method/Property | Description |
| :--- | :--- |
| `append()` | Concatenate with additional Index objects, producing a new Index |
| `difference()` | Compute set difference as an Index |
| `intersection()` | Compute set intersection |
| `union()` | Compute set union |
| `isin()` | Compute Boolean array indicating whether each value is contained in the passed collection |
| `delete()` | Compute new Index with element at Index $i$ deleted |
| `drop()` | Compute new Index by deleting passed values |
| `insert()` | Compute new Index by inserting element at Index $i$ |
| `is_monotonic` | Returns `True` if each element is greater than or equal to the previous element |
| `is_unique` | Returns `True` if the Index has no duplicate values |
| `unique()` | Compute the array of unique values in the Index |
