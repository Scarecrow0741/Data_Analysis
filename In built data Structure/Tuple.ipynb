# Tuple

### A tuple is a fixed-length, immutable sequence of Python objects which, once assigned,cannot be changed.


```python
tup = ( 1 , 2 , 3 )
tup
```

```python
tup = 1,2,3
tup  
```

```python
tup = tuple("String")
tup
```

```python
tup[0]
```

```python
nested_tup = ( (1,2,3), (4,5))
nested_tup
```

```python
tup = ( "foo" , [1,2] , True )
tup[1].append(3)
tup
```

```python
tup = (1,2,3) + ("foo",)
tup
```

```python
tup = (1,2) * 3
tup
```

```python
tup = (1,2,3)
a,b,c = tup
b
```

```python
seq = [ (1,2,3) ,( 4,5,6 )]
for a,b,c in seq:
    print(f"a is {a} b is {b} c is {c}")
```

```python
tup = 1,2,3,5,6
a,b,*something = tup # "*" holds the rest of the elements in the tuple 
print(a)
print(b)
print(something)
```

```python
tup = (1,2,2,3,2,5)
tup.count(2) # gives the number of occurences in the tuple 
```
