# Dictionary

### A dictionary stores a collection of key-value pairs, where key and value are Python objects. Each key is associated with a value so that a value can be conveniently retrieved, inserted, modified, or deleted given a particular key


```python
d1 = {"a": "some value", "b": [1, 2, 3, 4]}
d1
```

```python
d1[7] = "Something"
d1
```

```python
d1[7]
```

```python
d1[5] = "some value"
d1
```

```python
del d1[5]
d1
```

```python
ret = d1.pop(7)
ret
```

```python
words = ["apple", "bat", "bar", "atom", "book"]

by_letter = {}

for word in words:
    letter = word[0] 
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
        
#for word in words:
#    letter = word[0]
#    by_letter.setdefault(letter, []).append(word) # does the same thing 
        
by_letter 
```
