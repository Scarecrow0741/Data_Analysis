# List 

### In contrast with tuples, lists are variable length and their contents can be modified in place. Lists are mutable

```python
List = [1,2,3]
List 
```

```python
tup = ( 1,2,3 )
tup = list(tup)
tup
```

```python
list = [ "ok" , "gone" ]
list.append("let") #adds to the list
list
```

```python
list.insert(1,"in") #adds at a specific position 
list
```

```python
list.pop(1) # deletes element at a certain position
list.remove("let") # removes a specific element 
list
```

```python
list = [4, None, "foo"] + [7, 8, (2, 3)]
list
```

```python
list.extend( [7, 8, (2, 3)])
list
```

```python
a = [7, 2, 5, 1, 3]
a.sort()
a
```

```python
b = ["saw", "small", "He", "foxes", "six"]
b.sort(key=len) # sorting by length
b
```

```python
seq = [3,5,1,7,3,4]
seq[1:3]
```

```python
seq[1:3] = [6,7]
seq
```

```python
seq[::2]
```

```python
seq[::-1] #reverses the list 
```
