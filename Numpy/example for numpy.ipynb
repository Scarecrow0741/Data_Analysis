# Example: Random Walks

### first with just python code 

```python
import random 
position = 0 
walk = [position]
nsteps = 1000

for _ in range(nsteps):
    step = 1 if random.randint( 0,1 ) else -1 
    position += step
    walk.append(position)
```

```python
walk
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(walk[:100])
```

### now using numpy

```python
import numpy as np
```

```python
nsteps = 1000 
rng = np.random.default_rng(seed = 12345) #random generator
```

```python
draws = rng.integers(0,2,nsteps)
steps = np.where( draws == 0 , 1 , -1 )
walk = np.cumsum(steps)
walk
```

```python
plt.plot(walk[:100])
```

```python
walk.min() # gets the minimum value along the walks trajectory
```

```python
walk.max() #gets the maximum value along the walks trajectory
```

```python
(np.abs(walk) >=10).argmax() #gets the first time the walk reaches 10 or -10
```
