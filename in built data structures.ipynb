# Tuple

### A tuple is a fixed-length, immutable sequence of Python objects which, once assigned,cannot be changed.


```python
tup = ( 1 , 2 , 3 )
tup
```

```python
tup = 1,2,3
tup  
```

```python
tup = tuple("String")
tup
```

```python
tup[0]
```

```python
nested_tup = ( (1,2,3), (4,5))
nested_tup
```

```python
tup = ( "foo" , [1,2] , True )
tup[1].append(3)
tup
```

```python
tup = (1,2,3) + ("foo",)
tup
```

```python
tup = (1,2) * 3
tup
```

```python
tup = (1,2,3)
a,b,c = tup
b
```

```python
seq = [ (1,2,3) ,( 4,5,6 )]
for a,b,c in seq:
    print(f"a is {a} b is {b} c is {c}")
```

```python
tup = 1,2,3,5,6
a,b,*something = tup # "*" holds the rest of the elements in the tuple 
print(a)
print(b)
print(something)
```

```python
tup = (1,2,2,3,2,5)
tup.count(2) # gives the number of occurences in the tuple 
```

# List 

### In contrast with tuples, lists are variable length and their contents can be modified in place. Lists are mutable

```python
List = [1,2,3]
List 
```

```python
tup = ( 1,2,3 )
tup = list(tup)
tup
```

```python
list = [ "ok" , "gone" ]
list.append("let") #adds to the list
list
```

```python
list.insert(1,"in") #adds at a specific position 
list
```

```python
list.pop(1) # deletes element at a certain position
list.remove("let") # removes a specific element 
list
```

```python
list = [4, None, "foo"] + [7, 8, (2, 3)]
list
```

```python
list.extend( [7, 8, (2, 3)])
list
```

```python
a = [7, 2, 5, 1, 3]
a.sort()
a
```

```python
b = ["saw", "small", "He", "foxes", "six"]
b.sort(key=len) # sorting by length
b
```

# set

### A set is an unordered collection of unique elements

```python
s = set([1,2,3,2,4,4,6,6])
s
```

```python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7, 8} # or  a | b
a.union(b)
```

```python
a.intersection(b) # or a & b
a
```

### Table 3-1. Python set operations

| Function | Alternative syntax | Description |
| :--- | :--- | :--- |
| `a.add(x)` | N/A | Add element x to set a |
| `a.clear()` | N/A | Reset set a to an empty state, discarding all of its elements |
| `a.remove(x)` | N/A | Remove element x from set a |
| `a.pop()` | N/A | Remove an arbitrary element from set a, raising KeyError if the set is empty |
| `a.union(b)` | `a \| b` | All of the unique elements in a and b |
| `a.update(b)` | `a \|= b` | Set the contents of a to be the union of the elements in a and b |
| `a.intersection(b)` | `a & b` | All of the elements in *both* a and b |
| `a.intersection_update(b)` | `a &= b` | Set the contents of a to be the intersection of the elements in a and b |
| `a.difference(b)` | `a - b` | The elements in a that are not in b |
| `a.difference_update(b)` | `a -= b` | Set a to the elements in a that are not in b |
| `a.symmetric_difference(b)` | `a ^ b` | All of the elements in either a or b but *not both* |
| `a.symmetric_difference_update(b)` | `a ^= b` | Set a to contain the elements in either a or b but not both |
| `a.issubset(b)` | `<=` | True if the elements of a are all contained in b |
| `a.issuperset(b)` | `>=` | True if the elements of b are all contained in a |
| `a.isdisjoint(b)` | N/A | True if a and b have no elements in common |
